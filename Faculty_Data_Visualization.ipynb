# Data Acquisition

```python
import csv

# Define the name of the CSV file to save the data
csv_file_name = 'faculty_data.csv'

# Initialize the CSV file and write the header row
with open(csv_file_name, 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['Faculty Name', 'Department', 'Position', 'Email'])
```

```python
from bs4 import BeautifulSoup
import requests
import re

# Read HTML content from a text file
with open('fac1.txt', 'r', encoding='utf-8') as file:
    html_code = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_code, 'html.parser')

# Find all faculty items
faculty_items = soup.find_all('div', class_='faculty-item')
with open(csv_file_name, 'a', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)

    for faculty_item in faculty_items:
    # Extract faculty information
      faculty_name = faculty_item.find('p', class_='facul-title').find('a').text
      department = faculty_item.find('span', class_='yellow-tag').find('a').text
      position = faculty_item.find('p', class_='facul-desp').text
      position_text = faculty_item.find('p', class_='facul-desp').text
      position = re.sub(r'\s*(Dept\.|Department).*', '', position_text)
    # Find the URL of the faculty member's website
      faculty_website_url = faculty_item.find('a', href=True)['href']

    # Fetch and parse the content of the faculty member's website
      faculty_website_response = requests.get(faculty_website_url)
      faculty_website_soup = BeautifulSoup(faculty_website_response.text, 'html.parser')

    # Extract email from the faculty member's website
      email_link = faculty_website_soup.select_one('a[href^=mailto]')
      if email_link:
          email = email_link['href'].replace('mailto:', '')
      else:
          email = "N/A"

      print("Faculty Name:", faculty_name)
      print("Department:", department)
      print("Position:", position)
      print("Email:", email)
      print("\n")
      csv_writer.writerow([faculty_name, department, position, email])
```

# Data Visualization

```python
import matplotlib.pyplot as plt

department_counts = data['Department'].value_counts()

# Plotting a bar graph
department_counts.plot(kind='bar', color='skyblue')
plt.title('Number of People in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of People')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))

# Count the occurrences of each position
position_counts = data['Position'].value_counts()

# Plotting a horizontal bar chart
sns.barplot(x=position_counts, y=position_counts.index, palette='viridis')

plt.title('Distribution of Positions in the College')
plt.xlabel('Count')
plt.ylabel('Position')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")

departments = data['Department'].unique()

# Set up subplots based on the number of departments
fig, axes = plt.subplots(nrows=len(departments), ncols=1, figsize=(10, 4 * len(departments)))

for i, department in enumerate(departments):
    # Filter data for the current department
    dept_data = data[data['Department'] == department]

    # Count the occurrences of each position in the current department
    dept_pos_counts = dept_data['Position'].value_counts()

    # Plotting a horizontal bar graph for the current department
    sns.barplot(x=dept_pos_counts.values, y=dept_pos_counts.index, ax=axes[i], palette='viridis')
    axes[i].set_title(f'Distribution of Positions in {department}')

plt.tight_layout()
plt.show()
```
